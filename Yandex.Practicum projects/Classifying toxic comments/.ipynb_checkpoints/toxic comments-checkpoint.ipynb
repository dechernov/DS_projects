# Project for "Wikishop"

## Project description

**Brief information:** the online store "Wikishop" launches a new service. From now on, users can edit and add to product descriptions in the same way as in wiki communities. This means that customers can offer their own changes and comment on the changes made by others. The store needs a tool to find and moderate toxic comments. 

**Objective:** develop a machine learning model to classify comments as positive or negative.

**Tasks:** develop several machine learning models to classify comments into positive and negative with a quality metric value (F1) on the test sample of at least 0.75.

## Data description

The dataset contains a table with the following columns:
* `text` - column with the comment text;
* `toxic` - column with the target values (is the comment toxic or not).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb
import optuna
import re
import spacy
import swifter

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import f1_score, roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.utils import shuffle
from wordcloud import WordCloud
from optuna.samplers import TPESampler
```

```python
data = pd.read_csv('https://code.s3.yandex.net/datasets/toxic_comments.csv')
```

```python
data.head()
```

It looks like the column `Unnamed:0` represents an index. We'll delete this column and change the comments to lowercase.

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data['text'] = data['text'].str.lower()
```

```python
data.info()
```

```python
data.isna().mean()
```

```python
print(f'Number of duplicates in the dataframe: {data.duplicated().sum()}')
```

There are no missing values in the dataset, but the number of duplicates is 45 comments. Let's drop the duplicate values and transform our texts using the spaCY library, deleting all unnecessary symbols and keeping only letters.

```python
data.drop_duplicates(inplace=True)
```

```python
#loading spacy and stop words
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
stopwords = set(nlp.Defaults.stop_words)
```

```python
#lemmatization function and cleaning of text from symbols
def processing(text):    
    tweet = nlp(text)
    processed = " ".join([token.lemma_ for token in tweet])
    cleared = re.sub(r"[^a-z]", ' ', processed) 
    return " ".join(cleared.split())
```

```python
%%time
data['text_processed'] = data['text'].swifter.apply(processing)
```

## Building models

### Creating features

```python
features = data['text_processed']
target = data['toxic']
```

```python
#splitting the dataset
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=42, stratify=target)
```

```python
target_test.value_counts(normalize=True)
```

```python
target_train.value_counts(normalize=True)
```

### Dealing with imbalance (downsampling)

```python
#balancing classes function with downsampling
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=42)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=42)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=42)
    
    return features_downsampled, target_downsampled
```

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.1)

print(features_downsampled.shape)
print(target_downsampled.shape)
```

```python
#checking balance
target_downsampled.value_counts().plot(kind='bar', figsize=(15,5))
plt.title('Class distribution')
plt.xlabel('Class')
plt.ylabel('Number of values');
```

#### LogisticRegression

```python
%%time
def lr_downsampled(trial):    
    #pipeline containing TFIDF with stop-words and the model
    pipe_lr = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words=stopwords)),
        ('lr', LogisticRegression(random_state=42, max_iter=1000, n_jobs=-1))]
        )
    #params grid
    tfidf__min_df = trial.suggest_int('tfidf__min_df', 1, 10) 
    lr__solver = trial.suggest_categorical('lr__solver', ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'])
    lr__C = trial.suggest_loguniform('lr__C', 0.01, 10) 
    
    params = {
        'tfidf__min_df': tfidf__min_df,
        'lr__solver': lr__solver,
        'lr__C': lr__C
    }
    
    pipe_lr.set_params(**params)

    score = cross_val_score(pipe_lr, features_downsampled, target_downsampled, cv=3, n_jobs=-1, scoring='f1').mean()

    return score

sampler = TPESampler(seed=42) #sampler for repeatability
optuna.logging.set_verbosity(optuna.logging.INFO)
study_lr_down = optuna.create_study(direction='maximize', sampler=sampler)
study_lr_down.optimize(lr_downsampled, n_trials=15)
```

#### LGBM

```python
%%time
def lgb_downsampled(trial):    
    #pipeline containing TFIDF with stop-words and the model
    pipe_lgb = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words=stopwords)),
        ('lgbm', LGBMClassifier())]
        )
    #params grid
    tfidf__min_df = trial.suggest_int('tfidf__min_df', 1, 10)
    lgbm__max_depth = trial.suggest_int('lgbm__max_depth', 1, 10)
    lgbm__num_leaves = trial.suggest_int('lgbm__num_leaves', 1, 500)
    lgbm__learning_rate = trial.suggest_float('lgbm__learning_rate', 0.01, 0.1)
    lgbm__random_state = 42 
    lgbm__n_jobs = -1
    lgbm__boosting_type = 'gbdt'
    lgbm__objective = 'binary'
    lgbm__metric = 'binary_logloss'

    params = {
        'tfidf__min_df': tfidf__min_df,
        'lgbm__max_depth': lgbm__max_depth,
        'lgbm__num_leaves': lgbm__num_leaves,
        'lgbm__learning_rate': lgbm__learning_rate,
        'lgbm__random_state': lgbm__random_state,
        'lgbm__n_jobs': lgbm__n_jobs,
        'lgbm__bosting_type': lgbm__boosting_type,
        'lgbm__objective': lgbm__objective,
        'lgbm__metric': lgbm__metric
    }
    
    pipe_lgb.set_params(**params)

    score = cross_val_score(pipe_lgb, features_downsampled, target_downsampled, cv=3, n_jobs=-1, scoring='f1').mean()

    return score

sampler = TPESampler(seed=42) #sampler for repeatability
optuna.logging.set_verbosity(optuna.logging.INFO)
study_lgb_down = optuna.create_study(direction='maximize', sampler=sampler)
study_lgb_down.optimize(lgb_downsampled, n_trials=15)
```

#### RandomForest

```python
%%time
def rf_downsampled(trial):    
    #pipeline containing TFIDF with stop-words and the model
    pipe_rf = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words=stopwords)),
        ('rf', RandomForestClassifier(random_state=42, n_jobs=-1))]
        )
    #params grid
    tfidf__min_df = trial.suggest_int('tfidf__min_df', 1, 10) 
    rf__max_depth = trial.suggest_int('rf__max_depth',  2, 20)
    rf__n_estimators = trial.suggest_int('rf__n_estimators',  10, 350, 20)
    
    params = {
        'tfidf__min_df': tfidf__min_df,
        'rf__max_depth': rf__max_depth,
        'rf__n_estimators': rf__n_estimators
 
    }
    
    pipe_rf.set_params(**params)

    score = cross_val_score(pipe_rf, features_downsampled, target_downsampled, cv=3, n_jobs=-1, scoring='f1').mean()

    return score

sampler = TPESampler(seed=42) #sampler for repeatability
optuna.logging.set_verbosity(optuna.logging.INFO)
study_rf_down = optuna.create_study(direction='maximize', sampler=sampler)
study_rf_down.optimize(rf_downsampled, n_trials=15)
```

### Dealing with imbalance (balancing with internal methods)

#### LogisticRegression

```python
%%time
def lr_balanced(trial):    
    #pipeline containing TFIDF with stop-words and the model
    pipe_lr_bal = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words=stopwords)),
        ('lr', LogisticRegression(random_state=42, max_iter=1000, n_jobs=-1))]
        )
    #params grid
    tfidf__min_df = trial.suggest_int('tfidf__min_df', 1, 10) 
    lr__solver = trial.suggest_categorical('lr__solver', ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'])
    lr__C = trial.suggest_loguniform('lr__C', 0.01, 10) 
    lr__class_weight = 'balanced'
    
    params = {
        'tfidf__min_df': tfidf__min_df,
        'lr__solver': lr__solver,
        'lr__class_weight': lr__class_weight,
        'lr__C': lr__C
    }
    
    pipe_lr_bal.set_params(**params)

    score = cross_val_score(pipe_lr_bal, features_train, target_train, cv=3, n_jobs=-1, scoring='f1').mean()

    return score

sampler = TPESampler(seed=42) #sampler for repeatability
optuna.logging.set_verbosity(optuna.logging.INFO)
study_lr_bal = optuna.create_study(direction='maximize', sampler=sampler)
study_lr_bal.optimize(lr_balanced, n_trials=15)
```

#### LGBM

```python
%%time
def lgb_balanced(trial):    
    #pipeline containing TFIDF with stop-words and the model
    pipe_lgb_bal = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words=stopwords)),
        ('lgbm', LGBMClassifier())]
        )
    #params grid
    tfidf__min_df = trial.suggest_int('tfidf__min_df', 1, 10) 
    lgbm__max_depth = trial.suggest_int('lgbm__max_depth', 1, 10)
    lgbm__num_leaves = trial.suggest_int('lgbm__num_leaves', 1, 500)
    lgbm__learning_rate = trial.suggest_float('lgbm__learning_rate', 0.01, 0.1)
    lgbm__random_state = 42 
    lgbm__n_jobs = -1
    lgbm__boosting_type = 'gbdt'
    lgbm__objective = 'binary'
    lgbm__metric = 'binary_logloss'
    lgbm__class_weight = 'balanced'
        

    params = {
        'tfidf__min_df': tfidf__min_df,
        'lgbm__max_depth': lgbm__max_depth,
        'lgbm__num_leaves': lgbm__num_leaves,
        'lgbm__learning_rate': lgbm__learning_rate,
        'lgbm__random_state': lgbm__random_state,
        'lgbm__n_jobs': lgbm__n_jobs,
        'lgbm__bosting_type': lgbm__boosting_type,
        'lgbm__objective': lgbm__objective,
        'lgbm__metric': lgbm__metric,
        'lgbm__class_weight': lgbm__class_weight
    }
    
    pipe_lgb_bal.set_params(**params)

    score = cross_val_score(pipe_lgb_bal, features_train, target_train, cv=3, n_jobs=-1, scoring='f1').mean()

    return score

sampler = TPESampler(seed=42) #sampler for repeatability
optuna.logging.set_verbosity(optuna.logging.INFO)
study_lgb_bal = optuna.create_study(direction='maximize', sampler=sampler)
study_lgb_bal.optimize(lgb_balanced, n_trials=15)
```

#### RandomForest

```python
%%time
def rf_balanced(trial):    
    #pipeline containing TFIDF with stop-words and the model
    pipe_rf = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words=stopwords)),
        ('rf', RandomForestClassifier(random_state=42, n_jobs=-1))]
        )
    #params grid
    tfidf__min_df = trial.suggest_int('tfidf__min_df', 1, 10) 
    rf__max_depth = trial.suggest_int('rf__max_depth',  2, 20)
    rf__n_estimators = trial.suggest_int('rf__n_estimators',  10, 350, 20)
    rf__class_weight = 'balanced'
    
    params = {
        'tfidf__min_df': tfidf__min_df,
        'rf__max_depth': rf__max_depth,
        'rf__n_estimators': rf__n_estimators,
        'rf__class_weight': rf__class_weight
 
    }
    
    pipe_rf.set_params(**params)

    score = cross_val_score(pipe_rf, features_train, target_train, cv=3, n_jobs=-1, scoring='f1').mean()

    return score

sampler = TPESampler(seed=42) #sampler for repeatability
optuna.logging.set_verbosity(optuna.logging.INFO)
study_rf_bal = optuna.create_study(direction='maximize', sampler=sampler)
study_rf_bal.optimize(rf_balanced, n_trials=15)
```

### Conclusions

* **The best model is a logistic regression model trained on a downsampled dataset. The model parameters are: tfidf__min_df = 4, lr__solver = 'sag', lr__C = 9.133995846860973. However, since this model was trained on a balanced sample its results on the test sample may be significantly worse due to significant class imbalance.**
    
    **In this case, we will select the second best model for testing, but trained on the full sample - a logistic regression with the following hyperparameters: tfidf__min_df = 1, lr__solver = 'sag', lr__C = 3.063462210622083, lr__class_weight = 'balanced'. Let's check these models on the test sample.**

```python
pipe_lr_optimised = Pipeline([
        ('tfidf', TfidfVectorizer(min_df=study_lr_bal.best_params['tfidf__min_df'], stop_words=stopwords)),
        ('lr', LogisticRegression(random_state=42, max_iter=1000, 
                                  solver=study_lr_bal.best_params['lr__solver'], 
                                  C=study_lr_bal.best_params['lr__C'], n_jobs=-1,
                                 class_weight='balanced'))]
        )
```

```python
pipe_lr_optimised.fit(features_train, target_train)
preds_test = pipe_lr_optimised.predict(features_test)
```

```python
print(f'F1 metrics on the full set: {f1_score(target_test, preds_test)}')
```
