# Predicting steel melting temperature

## Project description

**Brief information**: In order to optimize production costs, the steel company decided to reduce power consumption in the steel processing stage. 

**Objective**: develop a model that predicts the temperature of the steel at the last measurement for the batch.

## Steel alloying process

## Data description

The data consists of files from different sources:

- `data_arc.csv` — electrode data;
- `data_bulk.csv` — data on bulk materials feed (volume);
- `data_bulk_time.csv` — data on bulk materials feed (time);
- `data_gas.csv` — data on alloy purging with gas;
- `data_temp.csv` — temperature measurement results;
- `data_wire.csv` — data on wire materials feed (volume);
- `data_wire_time.csv` — data on wire materials feed (time).

In all files, the `key` column contains the batch number. There may be several lines in the files with the same `key` value: they correspond to different processing iterations.

```python
#function for calculating the shares of each alloy 
def alloy_perc_calc(data):
    #total mass of alloys
    data['total_alloy'] = data['total_bulk'] + data['total_wire']
    
    #calculating proportion of each alloy
    for column in data.iloc[:, 1:].columns:
        data['{}_perc'.format(column)] = round((data[column] / data['total_alloy']) * 100, 2)
    
    #droping columns with absolute volumes
    data.drop(columns=['total_bulk', 'total_wire', 'total_alloy',
                      'bulk_1', 'bulk_2', 'bulk_3', 'bulk_4', 'bulk_5', 'bulk_6',
                      'bulk_7', 'bulk_8', 'bulk_9', 'bulk_10', 'bulk_11', 'bulk_12',
                      'bulk_13', 'bulk_14', 'bulk_15', 'wire_1', 'wire_2', 'wire_3', 'wire_4',
                      'wire_5', 'wire_6', 'wire_7', 'wire_8', 'wire_9', 'total_wire_perc',
                      'total_alloy_perc', 'total_bulk_perc'], axis=1, inplace=True)
```

## Building models

### Splitting dataset

```python
features = data.drop('last_temp', axis=1)
target = data['last_temp']

features_train, features_test, target_train, target_test = train_test_split(
                                                            features, 
                                                            target, 
                                                            test_size=0.25, 
                                                            random_state=RANDOM_STATE)
```

```python
features_lr = data_lr.drop('last_temp', axis=1)
target_lr = data_lr['last_temp']

features_train_lr, features_test_lr, target_train_lr, target_test_lr = train_test_split(
                                                            features_lr, 
                                                            target_lr, 
                                                            test_size=0.25, 
                                                            random_state=RANDOM_STATE)
```

```python
CV_FOLDS = 5
```

### Linear regression (ElasticNet)

```python
%%time
def lr(trial):    
    #pipeline with scaler and a model
    pipe_lr = Pipeline([
        ('scaler', StandardScaler()),
        ('lr', ElasticNet(random_state=RANDOM_STATE, max_iter=1000))]
        )
    
    #params grid
    
    lr__l1_ratio = trial.suggest_loguniform('lr__l1_ratio', 0.01, 1) 
    lr__selection = trial.suggest_categorical('lr__selection', ['cyclic', 'random'])
    
    params = {
        
        'lr__l1_ratio': lr__l1_ratio,
        'lr__selection': lr__selection
    }
    
    pipe_lr.set_params(**params)

    scores = cross_val_score(pipe_lr, features_train_lr, target_train_lr, cv=CV_FOLDS, n_jobs=-1, scoring='neg_mean_absolute_error')

    return scores.mean() * -1

sampler = TPESampler(seed=RANDOM_STATE) #optuna fixed sampler
optuna.logging.set_verbosity(optuna.logging.INFO)
study_lr = optuna.create_study(direction='minimize', sampler=sampler)
study_lr.optimize(lr, n_trials=30)
```

### LGBM

```python
%%time
#model creating function with optuna tuning
def objective_lgb(trial):
    
    param = { 
        'objective': 'regression',
        'random_state': RANDOM_STATE,
        'boosting_type': 'gbdt',
        'n_jobs': -1,
        'max_depth': trial.suggest_int('max_depth', 1, 10),
        'num_leaves' : trial.suggest_int('num_leaves', 1, 500),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1)
    }
    model = LGBMRegressor(**param) 
    
    scores = cross_val_score(model, features_train, target_train, scoring='neg_mean_absolute_error', cv=CV_FOLDS) 
   
    return scores.mean() * -1

sampler = TPESampler(seed=RANDOM_STATE) #optuna fixed sampler
optuna.logging.set_verbosity(optuna.logging.INFO)
study_lgb = optuna.create_study(direction='minimize', sampler=sampler)
study_lgb.optimize(objective_lgb, n_trials=30)

#best model
optimised_lgb = LGBMRegressor(max_depth=study_lgb.best_params['max_depth'], num_leaves=study_lgb.best_params['num_leaves'],
                                     learning_rate = study_lgb.best_params['learning_rate'], boosting_type='gbdt', random_state=RANDOM_STATE, n_jobs=-1, 
                                     objective='regression')

#fitting best model
optimised_lgb.fit(features_train, target_train)
```

### CatBoost

```python
%%time
#model creating function with optuna tuning
def objective_cb(trial):

    param = {
       'loss_function': 'RMSE',
       'learning_rate': trial.suggest_float('learning_rate', 0.01,0.1),
       'n_estimators': trial.suggest_int( "n_estimators",  40,300,20),
       'max_depth': trial.suggest_int( 'max_depth',  1,8,1),
       'random_seed': RANDOM_STATE
       }
    
    model = CatBoostRegressor(**param, verbose=150)  
    
    scores = cross_val_score(model, features_train, target_train, scoring='neg_mean_absolute_error', cv=CV_FOLDS) 
   
    return scores.mean() * -1

sampler = TPESampler(seed=RANDOM_STATE) #optuna fixed sampler
study_cb = optuna.create_study(direction='minimize', sampler=sampler)
study_cb.optimize(objective_cb, n_trials=30)

#best model
optimised_cb = CatBoostRegressor(max_depth=study_cb.best_params['max_depth'], learning_rate=study_cb.best_params['learning_rate'],
                                     n_estimators = study_cb.best_params['n_estimators'], random_seed=RANDOM_STATE, 
                                     loss_function='RMSE')

#fitting best model
optimised_cb.fit(features_train, target_train)
```

### RandomForest

```python
%%time
#model creating function with optuna tuning
def objective_rf(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 350),
        'max_depth': trial.suggest_int('max_depth', 4, 20),
        'max_features': trial.suggest_int('max_features', 2, 9)
        }

        
    model = RandomForestRegressor(random_state=RANDOM_STATE, **params, n_jobs=-1)
        
    scores = cross_val_score(model, features_train, target_train, scoring='neg_mean_absolute_error', cv=CV_FOLDS)

    return scores.mean() * -1

sampler = TPESampler(seed=RANDOM_STATE) #optuna fixed sampler
study_rf = optuna.create_study(direction='minimize', sampler=sampler)
study_rf.optimize(objective_rf, n_trials=30)

#best model
optimised_rf = RandomForestRegressor(max_depth=study_rf.best_params['max_depth'], n_estimators=study_rf.best_params['n_estimators'],
                                     max_features = study_rf.best_params['max_features'], random_state=RANDOM_STATE)

#fitting best model
optimised_rf.fit(features_train, target_train)
```

To save energy and production costs, it is therefore important to control the temperature when the steel arrives for the alloying process. For example, the temperature can be influenced by the speed at which the ladle arrives at the plant after the previous alloying process or by the time it spends in the processing queue.


In fact, this model is the primary step in analyzing the influence of factors on the final alloy temperature. Other processes that precede alloying should be further considered and the possibility of influencing and controlling the temperature during such processes should be evaluated.
