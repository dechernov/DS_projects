```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
place_name = ['Moscow', 'Moscow Oblast']
tags = {'railway': 'station', 'railway': 'halt'}
stations = ox.geometries_from_place(place_name, tags)
stations.head()
```

```python
import pandas as pd

st = pd.DataFrame(data=stations,columns=['express:user','name', 'geometry'])
```

```python
st.reset_index().to_csv('stations.csv',sep=',')
```
